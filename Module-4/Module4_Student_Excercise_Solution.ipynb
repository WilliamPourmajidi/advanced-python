## 1. Load and Inspect the Penguins Dataset

```python
import pandas as pd
import seaborn as sns
import numpy as np

df = sns.load_dataset('penguins')
df.head()
```

## 2. Dataset Overview
**Note:** 344 rows, 7 columns. Numeric: bill_length_mm, bill_depth_mm, flipper_length_mm, body_mass_g. Categorical: species, island, sex.

```python
df.shape
```

```python
df.info()
```

## 3. Descriptive Statistics

```python
df.describe(include='all')
```

```python
col = 'body_mass_g'
print("Mean body mass (g):", df[col].mean())
print("Median body mass (g):", df[col].median())
print("Mode body mass (g):", df[col].mode()[0])
```

## 4. Handling Missing Values
**Note:** body_mass_g has 2 missing values; sex has 11. The numeric measurement columns each have 2. This is a good moment to point out that `isnull().sum()` is the fastest first scan for data quality.

```python
df.isnull().sum()
```

### 4a. Drop

```python
df_clean = df.dropna(subset=['body_mass_g'])
print(f"Original dataset size: {df.shape[0]} rows")
print(f"After dropping missing 'body_mass_g': {df_clean.shape[0]} rows")
print(f"Rows removed: {df.shape[0] - df_clean.shape[0]}")
```

### 4b. Impute
**Note:** Median is the safer default for a skewed or outlier-prone column; mean is fine for roughly symmetric data. Body mass here is fairly symmetric, so mean and median land close together. This is a good discussion hook.

```python
df['body_mass_imputed'] = df['body_mass_g'].fillna(df['body_mass_g'].median())
df.head(10)
```

## 5. Measures of Spread

```python
col = 'body_mass_g'
print("Min:", df_clean[col].min())
print("Max:", df_clean[col].max())
print("Range:", df_clean[col].max() - df_clean[col].min())
print("IQR:", df_clean[col].quantile(0.75) - df_clean[col].quantile(0.25))
print("Variance:", df_clean[col].var())
print("Std Dev:", df_clean[col].std())
```

## 6. Correlation and Covariance
**Note:** Strongest correlation is flipper_length_mm vs body_mass_g (about +0.87): longer flippers go with heavier birds. bill_depth_mm is negatively correlated with the others, which surprises students and is worth unpacking (it is driven by species differences, an example of Simpson's paradox if you want to go deeper).

```python
df_clean.corr(numeric_only=True)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.heatmap(df_clean.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
df_clean.cov(numeric_only=True)
```

```python
sns.heatmap(df_clean.cov(numeric_only=True), annot=True, cmap='YlGnBu')
plt.title('Covariance Matrix Heatmap')
plt.show()
```

## 7. Stretch Challenges — sample solutions

```python
# 1. Mean flipper length per island
print(df.groupby('island')['flipper_length_mm'].mean(), "\n")

# 2. Mean body mass by species and sex
print(df.groupby(['species', 'sex'])['body_mass_g'].mean(), "\n")

# 3. Mean vs median imputation
df['body_mass_mean_imputed'] = df['body_mass_g'].fillna(df['body_mass_g'].mean())
print("Median fill:", df['body_mass_imputed'].mean())
print("Mean fill:  ", df['body_mass_mean_imputed'].mean(), "\n")

# 4. IQR outlier rule
q1 = df_clean['body_mass_g'].quantile(0.25)
q3 = df_clean['body_mass_g'].quantile(0.75)
iqr = q3 - q1
low, high = q1 - 1.5*iqr, q3 + 1.5*iqr
outliers = df_clean[(df_clean['body_mass_g'] < low) | (df_clean['body_mass_g'] > high)]
print(f"IQR outlier bounds: [{low:.0f}, {high:.0f}]")
print("Number of outliers:", len(outliers))
```
